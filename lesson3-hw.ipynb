Поменяйте его форму так, чтобы получилась квадратная матрица C. <br/>
Получите матрицу D, поэлементно прибавив матрицу B из предыдущего вопроса к матрице C, умноженной на 10.<br/>
Вычислите определитель, ранг и обратную матрицу D_inv для D.

```python
np.random.seed(42)
c = np.random.randint(0,16,16)
c
```

```python
C = c.reshape(4,4)
C
```

```python
D = B + C*10
D
```

```python
# Опеределитель матрицы D
D_det = np.linalg.det(D)
D_det
```

```python
# Ранг матрицы D
D_rank = np.linalg.matrix_rank(D)
D_rank
```

```python
# Обратная матрица D
D_inv = np.linalg.inv(D)
D_inv
```
